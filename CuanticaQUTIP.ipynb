# Graficando la esfera de Bloch

Para hacer este punto de diferenciar estados cuanticos de los qubits se va a hacer uso de la representación gráfica en la esfera de Bloch, para ello se usa Qutip, una libreria de Python que tiene ya implementada la visualización.A continuación se instala y se carga la esfera

```
!pip install qutip
```

```
import qutip
```

Esta es la esfera de Bloch vacia sin ningun vector de estado

```
b=qutip.Bloch()
b.show()
```

a)entre $|0> $ y $-|0>$ ya que la fase global no tiene relevancia física, es decir la funcion de onda no toma relevancia si no hasta que se toma su norma al cuadrado para obtener la probabilidad, estos estados son equivalentes.
ya que 
$$
|\psi\rangle=|0> \rightarrow \quad \alpha=1 \qquad  |\alpha|^2=1
\\
|\psi\rangle^{\prime}=-|0\rangle \rightarrow \quad \alpha=-1 \qquad  |\alpha|^2=1
$$
tendran la misma representación en la esfera de bloch.

```
vecb = [0,0,1]
b.add_vectors(vecb)
b.show()
```

b) para $|1\rangle$ y  $i|1\rangle$
 de nuevo las fases globales no tienen significado físico por lo que representan el mismo estado ya que 
 $$
 |\psi\rangle=|1\rangle \quad \rightarrow \beta=1 \qquad |\beta|^2=1
 \\
 |\psi\rangle=i|1\rangle \quad \rightarrow \beta=i \qquad |\beta|^2=1
 $$
 Su representación en la esfera de Bloch es la misma


```
veca = [0,0,-1]
a=qutip.Bloch()
a.add_vectors(veca)
a.show()
```

c)Esta vez los estados son difentes ya que la fase de diferencia no es global por lo que los estados son
$$
|\psi\rangle_1=\frac{1}{\sqrt{2}}\left(|0\rangle+|1\rangle\right)
 \\
 |\psi\rangle_2=\frac{1}{\sqrt{2}}\left(|0\rangle+i|1\rangle\right)
$$
En la siguiente figura el vector de color verde es $|\psi>_1$ y el  es $|\psi>_2$

```
vecc = [1,0,0]
vecd = [0,1,0]
c=qutip.Bloch()
c.add_vectors(vecc)
c.add_vectors(vecd)
c.show()
```

d)Para el siguiente caso la fase global es $e^{i\pi}$por lo que son el mismo estado 
$$ |\psi\rangle_1=\frac{1}{\sqrt{2}}(-|0\rangle+|1\rangle) \quad |\psi\rangle_2=\frac{1}{\sqrt{2}}(|0\rangle-|1\rangle)$$

```
vece = [0,-1,0]
e=qutip.Bloch()
e.add_vectors(vece)
e.show()
```

e) Para el ultimo caso se tiene
$$ |\psi\rangle_1=\frac{1}{\sqrt{2}}(|0\rangle+e^{i\pi/4}|1\rangle) \quad |\psi\rangle_2=\frac{1}{\sqrt{2}}(e^{-i\pi/4}|0\rangle+|1\rangle)$$
Vemos que al multiplicar por una fase global($e^{i\pi/4}$) a $|\psi\rangle_2$ de manera que :
$$e^{i\pi/4}\cdot|\psi\rangle_2=\frac{1}{\sqrt{2}}(e^{i\pi/4}e^{-i\pi/4}|0\rangle+e^{i\pi/4}|1\rangle)=\frac{1}{\sqrt{2}}(|0\rangle+e^{i\pi/4}|1\rangle) = |\psi\rangle_1$$
ya que a diferencia de una fase global son iguales, y las fases no son observables fisicos decimos que $|\psi\rangle_1=|\psi\rangle_2$
