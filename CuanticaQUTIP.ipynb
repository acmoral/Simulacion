# Graficando la esfera de Bloch

```
import qutip
```

```
b=qutip.Bloch()
b.show()
```

entre $|0> $ y $-|0>$ ya que la fase global no tiene relevancia física, es decir la funcion de onda no toma relevancia si no hasta que se toma su norma al cuadrado para obtener la probabilidad, estos estados son equivalentes.
ya que 
$$
|\psi>=|0> \rightarrow \quad \alpha=1 \qquad  |\alpha|^2=1
\\
|\psi>^{\prime}=-|0> \rightarrow \quad \alpha=-1 \qquad  |\alpha|^2=1
$$
tendran la misma representación en la esfera de bloch.

```
vec = [0,0,1]
b.add_vectors(vec)
b.show()
```

b) para $|1>$ y  $i|1>$
 de nuevo las fases globales no tienen significado físico por lo que representan el mismo estado ya que 
 \begin{gather*}
 |\psi>=|1> \quad \rightarrow \beta=1 \qquad |\beta|^2=1
 \\
 |\psi>=i|1> \quad \rightarrow \beta=i \qquad |\beta|^2=1
 \end{gather*}
 Su representación en la esfera de Bloch es la misma
 

```
vec = [0,0,-1]
a=qutip.Bloch()
a.add_vectors(vec)
a.show()
```

Esta vez los estados son difentes ya que la fase de diferencia no es global por lo que los estados son
\begin{gather*}
|\psi>_1=\frac{1}{\sqrt{2}}\left(|0>+|1>\right)
 \\
 |\psi>_2=\frac{1}{\sqrt{2}}\left(|0>+i|1>\right)
\end{gather*}
